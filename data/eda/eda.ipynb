```python
import numpy as np
variables_iniciales = ['tiempo_viaje_acumulado', 'carga_camion', 'num_localizaciones_visitadas',
                       'num_localizaciones_planificadas', 'tiempo_viaje_estimado',
                       'num_paquetes_entregados', 'num_letras_sim']

# Calcular la matriz de correlación
corr_matrix = df_features[variables_iniciales].corr().abs()

# Seleccionar la parte superior de la matriz de correlación (triangular) para evitar duplicados
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Mostrar variables altamente correlacionadas (umbral, por ejemplo, 0.9)
umbral = 0.9
print("Variables altamente correlacionadas (coeficiente > {}):".format(umbral))
print(upper_tri[upper_tri > umbral].dropna(how='all', axis=0))

# Suponiendo que encontramos que 'tiempo_viaje_estimado' y 'tiempo_viaje_acumulado' están muy correlacionados,
# podríamos decidir conservar solo 'tiempo_viaje_estimado'.
variables_seleccionadas = ['tiempo_viaje_estimado', 'carga_camion', 'num_localizaciones_visitadas',
                           'num_localizaciones_planificadas', 'num_paquetes_entregados', 'num_letras_sim']

# También se pueden crear variables dummy si se decide incluir la localización de destino
if 'destination' in df_features.columns:
    destination_dummies = pd.get_dummies(df_features['destination'], prefix='dest', drop_first=True)
    df_features_model = pd.concat([df_features[variables_seleccionadas], destination_dummies], axis=1)
else:
    df_features_model = df_features[variables_seleccionadas]

# Ahora df_features_model contiene las variables de entrada para el modelo de ML.
print("Variables de entrada para el modelo:")
df_features_model.head()
```

```python
import pandas as pd
import numpy as np

# Supongamos que ya tienes el dataframe 'df_features_model' con las variables:
# - tiempo_viaje_estimado
# - carga_camion
# - num_localizaciones_visitadas
# - num_localizaciones_planificadas
# - num_paquetes_entregados
# - num_letras_sim
# - Variables dummy para 'destination' (por ejemplo, columnas que empiezan con 'dest_')
# - (Adicionalmente pueden existir identificadores como simulationId o truck_id que descartaremos)

# Paso 1: Seleccionar las variables base (descartamos identificadores)
variables_base = ['tiempo_viaje_estimado', 'carga_camion', 'num_localizaciones_visitadas',
                  'num_localizaciones_planificadas', 'num_paquetes_entregados', 'num_letras_sim']

# Paso 2: Reducir la dimensionalidad de las dummies de 'destination'
# Asumimos que las variables dummy tienen un prefijo, por ejemplo, 'dest_'
dummy_cols = [col for col in df_features_model.columns if col.startswith('dest_')]

# Contar la frecuencia total de cada dummy (la suma de la columna indica cuántas veces aparece esa categoría)
dummy_frecuencias = df_features_model[dummy_cols].sum().sort_values(ascending=False)
# Seleccionar, por ejemplo, las 10 categorías principales
top_dummy_cols = dummy_frecuencias.head(10).index.tolist()

# Combinar las variables base con las dummies seleccionadas
variables_model = variables_base + top_dummy_cols

# Construir el dataframe final de features
df_model = df_features_model[variables_model]

# Paso 3: Eliminar variables altamente correlacionadas entre las variables numéricas
# Solo consideramos las variables numéricas (excluyendo las dummies de destino, que son 0/1)
df_numericas = df_model[variables_base]

# Calcular la matriz de correlación absoluta
corr_matrix = df_numericas.corr().abs()

# Tomamos la parte superior (triangular) para identificar pares redundantes
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Establecer un umbral (por ejemplo, 0.9) para considerar alta correlación
umbral = 0.9

# Identificar las variables que se pueden eliminar (si tienen alta correlación con alguna otra)
variables_a_eliminar = [col for col in upper_tri.columns if any(upper_tri[col] > umbral)]
print("Variables numéricas a eliminar por alta correlación:", variables_a_eliminar)

# Crear el conjunto final de variables numéricas seleccionadas (descartar las variables redundantes)
variables_numericas_final = [var for var in variables_base if var not in variables_a_eliminar]

# Combinar con las dummies seleccionadas (estas no se consideran para la correlación ya que son binarias)
variables_final = variables_numericas_final + top_dummy_cols
print("Variables finales seleccionadas para el modelo:")
print(variables_final)

# Dataframe final para el modelo
df_model_final = df_model[variables_final]
print("\nPrimeras filas del dataset final para el modelo:")
df_model_final.head()
```
